```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
rng = np.random.default_rng(2023)
```

```python
x = rng.normal()
y = rng.normal()
```

```python
plt.plot(x,y,'o',color='orange')
```

```python
# 创建多个堆叠的就需要用到subplot（）
fig , ax = plt.subplots(figsize=(8,8))
x = rng.normal()
y = rng.normal()
ax.plot(x,y,'o')
p = rng.normal()
q = rng.normal()
ax.plot(p,q,'x',color = 'orange')
```

```python
# 创建多个堆叠的就需要用到subplot（）
fig , ax = plt.subplots(figsize=(8,8))
x = rng.normal()
y = rng.normal()
ax.plot(x,y,'o')
p = rng.normal()
q = rng.normal()
ax.scatter(p,q,marker='x',color = 'orange')
ax.set_xlabel("x_label")
ax.set_ylabel("y_label")
ax.set_title("Plot title")
```

```python
#进入fig 意味着我们可以改变一些方面
fig.set_size_inches(12,3)
```

```python
#创建网格图

fig,axes = plt.subplots(nrows=2,ncols=3,figsize=(15,5))
```

```python
axes[0,1].plot(x,y,'o')
axes[1,2].scatter(x,y,marker="+")
fig
```

```python
#保存图片 dpi指定像素点

fig.savefig('Figure.png',dpi = 400)
```

```python
# 逐步更新figure

axes[0,1].set_xlim([-1,1]) #设置x轴的范围
fig
```

```python
# 绘制更复杂的图像

fig , ax = plt.subplots(figsize = (8,8))
x = np.linspace(-np.pi,np.pi,50)
y = x
f = np.multiply.outer(np.cos(y),1/(1+x**2))
ax.contour(x,y,f)
```

```python
ax.contour(x,y,f,levels = 5)
fig
contour=ax.contour(x,y,f,levels = 5)
```

```python
plt.clabel(contour,fontsize=10)
fig
```

```python
fig , ax = plt.subplots(figsize = (8,8))
x = np.linspace(-np.pi,np.pi,50)
y = x
f = np.multiply.outer(np.cos(y),1/(1+x**2))
ax.contourf(x,y,f)
```

```python
fig , ax = plt.subplots(figsize = (8,8))
x = np.linspace(-np.pi,np.pi,50)
y = x
f = np.multiply.outer(np.cos(y),1/(1+x**2))
im = ax.imshow(f)
fig.colorbar(im,ax=ax)
```

```python
# 继续优化 需要将x，y轴的值放入

fig , ax = plt.subplots(figsize = (8,8))
x = np.linspace(-np.pi,np.pi,50)
y = x
f = np.multiply.outer(np.cos(y),1/(1+x**2))
im = ax.imshow(f,extent=[x.min(),x.max(),y.min(),y.max()])
fig.colorbar(im,ax=ax,fraction=0.03, pad=0.05)
```
